```python
import pandas as pd
```

##### Create series with array

```python
students = ["Avi","Aditya","Molly"]
pd.Series(students)
```

```python
numbers =[1,2,3]
pd.Series(numbers)
```

```python
students = ["avi","Aditya",None]
pd.Series(students)
```

##### Create series with dictionary

```python
students_scores ={"Alice":"Physics",
                  "Jack":"Chemistry",
                  "Molly": "English"}
s =pd.Series(students_scores)
s
```

```python
s.index
```

```python
students =[("Alice","Brown"),("Jack","white"),("Molly","Green")]
pd.Series(students)
```

```python
s =pd.Series(["Physics","Chemistry","English"],index=["Alice","Jack","Molly"])
s
```

```python
students_scores ={"Alice":"Physics",
                  "Jack":"Chemistry",
                  "Molly": "English"}
s =pd.Series(students_scores,index=["Alice","Molly","Sam"])
s
```
