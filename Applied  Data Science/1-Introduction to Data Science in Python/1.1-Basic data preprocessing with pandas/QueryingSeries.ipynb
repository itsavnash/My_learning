### Querying Series

```python
import pandas as pd
```

```python
students_scores ={"Alice":"Physics",
                  "Jack":"Chemistry",
                  "Molly": "English",
                  "Sam":"History"}
s =pd.Series(students_scores)
s
```

#### iloc vs loc 


##### iloc (Index Location)

- Meaning: Used to access rows and columns by integer (numeric) position.

- Think of it as: “I want the item at this position (number)”
- Syntax:-
  - df.iloc[row_index, column_index]

```python
s.iloc[3]
```

##### loc (Label Location)

- Meaning: Used to access rows and columns by labels or names.

- Think of it as: “I want the item with this label/name”
- Syntax:-
  - df.loc[row_label, column_label]



```python
s.loc["Molly"]
```

```python
s[3]
```

```python
s["Molly"]
```
