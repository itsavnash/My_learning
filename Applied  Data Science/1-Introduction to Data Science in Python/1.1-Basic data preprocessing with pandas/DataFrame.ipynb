```python
import pandas as pd
```

```python
record1 = pd.Series({"Name": "Alice",
                     "Class": "Physics",
                     "Score":85})
record2 = pd.Series({"Name": "Jack",
                     "Class": "Chemistry",
                     "Score":82})
record3 = pd.Series({"Name": "Helan",
                     "Class": "Biology",
                     "Score":90})
```

```python
df =pd.DataFrame([record1,record2,record3],
                  index=["school1","school2","school3"])
df.head()
```
