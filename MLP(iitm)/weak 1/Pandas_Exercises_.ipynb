### 1.4

Create a new column 'Price_Category' that labels each sale as:
'Budget' if Unit_Price < 300
'Mid-Range' if 300  Unit_Price < 1000
'Premium' if Unit_Price  1000

```python
sales_data
```

```python
def bin(row):

  if row<300:
    return "Budget"

  elif row>=300 and row<1000:
    return "Mid-Range"

  else:
    return "Premium"


sales_data['Price_Category']  = sales_data['Unit_Price'].apply(bin)
sales_data
```
