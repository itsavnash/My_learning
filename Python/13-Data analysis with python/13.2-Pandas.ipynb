#### Pandas - DataFrame and Series

```python
import pandas as pd
```

##### Series

```python
data = [1,2,3,4,5]
series= pd.Series(data)
print("series \n",series)
print(type(series))
```

```python
## Creating  a Series from dictinary
data = {"A": 1, "B":2,"C":3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data  =[10,20,30]
index =["a","b","c"]
pd.Series(data,index=index)
```

#### DataFrame

```python
## create a dataframe for dict of list

data = {
    "Name": ["Avinash","Aditya","Rohit"],
    "Age":[25,35,20],
    "City": ["Bangalore","New york","Florida"]
    }
df =pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## create a Data frame from a list of dict

data = [
     {"Name": "Avinash", "Age":25,"City":"Bangalore"},
     {"Name": "Aditya", "Age":32,"City":"New york"},
     {"Name": "Rohit", "Age":25,"City":"Florida"}
    
]

df =pd.DataFrame(data)
print(df)
print(type(df))
```
