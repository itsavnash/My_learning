1 - Given a number n, check whether it is even or odd. Return true for even and false for odd.

```python
def is_even(n):
    if n % 2 ==0:
        return True
    else:
        return False
```

2 - Given a number n, we need to print its table. 

```python
def table(n):
    for i  in range(1,11):
        print ("%d * %d = %d" % (n, i, n * i))
```

```python
table(5)
```
